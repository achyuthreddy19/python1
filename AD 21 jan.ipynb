```python
import pandas as pd
dataset=pd.read_csv(R'C:\Users\91773\Downloads\tweets.csv', encoding='ISO-8859-1')
dataset.head()
import re
```

```python
def clean_text(text):
    text=re.sub(r'RT','',text)
    text=re.sub(r'&amp;','',text)
    text=re.sub(r'[?!.;:,#@-]','',text)
    return text
```
